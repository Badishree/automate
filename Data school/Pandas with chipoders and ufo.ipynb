```python
# When can we not use the df. notation for getting series.
# If the colomn name has a space in between like 'Colors reported'
# When the col_name is already an attribute like 'shape' or 'head'
# the df. notation has limitations but The df['col_name'] works everytime
```

```python
ufo.Location = ufo.City + ', ' + ufo.State
# this gives a warning
```

```python
ufo.head()
# we can see that the Location has not been appended in the df
```

```python
ufo['Location'] = ufo.City + ', ' + ufo.State
# Hence, we must use the df['col_name'] to create a column or series in the df 
```

```python
ufo.head()
# Now we can see the Location column in the df
```

```python
# How do I rename columns in pandas
ufo.columns
```

```python
# There are 3 ways.
# One is to rename with df.rename command

ufo.rename(columns = {'Colors Reported' : 'Colors_Reported' , 'Shape Reported' : 'Shape_Reported'} , inplace = True)
```

```python
ufo.columns
```

```python
# another way is to use col_names[] list 
# you can use this in the pd.read_csv command as well as an argument in the names = 'list' with header = 0,

ufo_colnames =['city', 'colors reported', 'shape reported', 'state', 'time', 'location']
```

```python
ufo.columns = ufo_colnames
```

```python
ufo.columns
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports', names = ufo_colnames, header = 0) # have to change to 5 names to run
```

```python
# 3rd way, the best way for large no. of columns, is to use str method.
# we can use various string methods with the series.str.method command

ufo.columns = ufo.columns.str.replace(' ','_')
```
